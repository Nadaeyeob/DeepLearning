```python
import os
import random

VOC_ROOT_DIR = 'C:/Program Files/DL/VOCtrainval_11-May-2012/VOCdevkit/VOC2012/'
ANNO_DIR = os.path.join(VOC_ROOT_DIR,'Annotations')
IMAGE_DIR = os.path.join(VOC_ROOT_DIR, 'JPEGImages')

xml_files = os.listdir(ANNO_DIR)
print(xml_files[:5])
print(len(xml_files))
```

```python
xml_file = os.path.join(ANNO_DIR,'2007_000032.xml')
tree = ET.parse(xml_file)
root = tree.getroot()
```

```python
image_name = root.find('filename').text
full_image_name = os.path.join(IMAGE_DIR, image_name)
image_size = root.find('size')
image_width = int(image_size.find('width').text)
image_height = int(image_size.find('height').text)
```

```python
objects_list = []
for obj in root.findall('object'):
    xmlbox = obj.find('bndbox')
    x1= int(xmlbox.find('xmin').text)
    y1= int(xmlbox.find('ymin').text)
    x2= int(xmlbox.find('xmax').text)
    y2= int(xmlbox.find('ymax').text)
    bndbox_pos = (x1,y1,x2,y2)
    class_name = obj.find('name').text
    object_dict = {'class_name' : class_name, 'bndbox_pos':bndbox_pos}
    objects_list.append(object_dict)
    
print(full_image_name,'\n', image_size)
```

```python
for object in objects_list:
    print(object)
```

### OPEN CV  ###

```python
import cv2
import os
import xml.etree.ElementTree as ET
```

```python
xml_file = os.path.join(ANNO_DIR,'2007_000032.xml')
tree = ET.parse(xml_file)
root = tree.getroot()
file_name = root.find('filename').text
file_full_name = os.path.join(IMAGE_DIR, file_name)
print(file_full_name)
```

```python
img = cv2.imread(file_full_name)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_copy = img_rgb.copy()
```

```python
# BGR  RGB      
green_color = (0,255,0)
red_color = (255,0,0)
```

```python
object_list = []
for obj in root.findall('object'):
    xmlbox = obj.find('bndbox')
    left = int(xmlbox.find('xmin').text)
    top = int(xmlbox.find('ymin').text)   
    right = int(xmlbox.find('xmax').text)    
    bot = int(xmlbox.find('ymax').text)
    
    class_name = obj.find('name').text
    
    cv2.rectangle(img_copy, (left,top),(right,bot),color = green_color, thickness = 1)
    cv2.putText(img_copy, class_name,(left,top-5), fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=0.4,color= red_color, thickness=1)
    
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img_copy)
plt.show()
```
