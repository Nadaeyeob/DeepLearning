### IOU 구하기 ###

```python
img = cv2.imread('D:/DL/DLCV-master/data/image/audrey01.jpg')

gt_box = [60,15,320,420]
red = (255,0,0)
img_copy = img.copy()
img_gt_box = cv2.rectangle(img,(60,15),(320,420),color = red, thickness=2)
plt.figure(figsize=(3,3))
plt.imshow(img_gt_box)
plt.show()
```

```python
def iou_eval(cand_box, gt_box):
    left_top_x = cand_box[0]
    left_top_y = cand_box[1]
    right_bot_x = left_top_x + cand_box[2]
    right_bot_y = left_top_y + cand_box[3]
    
    int_left_top_x = np.maximum(left_top_x,gt_box[0])
    int_left_top_y = np.maximum(left_top_y,gt_box[1])
    int_right_bot_x = np.minimum(right_bot_x,gt_box[2])
    int_right_bot_y = np.minimum(right_bot_y,gt_box[3])
    
    intersection = np.maximum(int_right_bot_x - int_left_top_x,0) * np.maximum(int_right_bot_y - int_left_top_y,0)
    gt_box_area = (gt_box[2] - gt_box[0]) * (gt_box[3] - gt_box[1])
    cand_box_area = (right_bot_x - left_top_x) * (right_bot_y - left_top_y)
    total_area = gt_box_area + cand_box_area - intersection
    
    iou = intersection / total_area
    return iou
```

```python
_,regions = selectivesearch.selective_search(img_copy,min_size=2000, scale=100)
```

```python
cand_boxs = [cand['rect'] for cand in regions if cand['size'] > 5000]
```

```python
gt_box = [60,15,320,420]
for i,cand_box in enumerate(cand_boxs):
    iou = iou_eval(cand_box,gt_box)
    print(i, iou)
```

```python
gt_box = [60,15,320,420]
for i,cand_box in enumerate(cand_boxs):
    iou = iou_eval(cand_box,gt_box)
    
    if iou > 0.5:
        img_iou = cv2.rectangle(img_gt_box,(cand_box[0],cand_box[1]),(cand_box[2],cand_box[3]),color=green_rgb, thickness = 2)
        print(i, iou)
        text = '{}:{:.2f}'.format(i,iou)
        cv2.putText(img_gt_box, text, (cand_box[0],cand_box[1]+10),cv2.FONT_HERSHEY_SIMPLEX, 0.4,color = green_rgb)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img_gt_box)
plt.show()
```
